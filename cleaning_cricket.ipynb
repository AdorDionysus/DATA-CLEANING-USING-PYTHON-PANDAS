```python
import pandas as pd
import matplotlib as plt
```

```python
#importing csv file to dataframe AND renaming columns

df = pd.read_csv("cricket_clean.csv")


df = df.rename(columns=
          {'Mat': 'Matches_Played', 'Inns' : 'Innings_Batted', 'NO' : 'Not_Outs', 'HS': 'Highest_Inns_Score',
           'BF' : 'Ball_Faced', 'SR': 'Batting_Strike_Rate'
          }
         ).copy()
```

```python
#CLEANING AND SPLITTING PLAYER COLUMN AND CREATE COUNTRY COLUMN

df['Country'] = df['Player'].str.split("(").str[1]
df['Country'] = df['Country'].str.split(")").str[0]
df['Player'] = df['Player'].str.split("(").str[0]
df
```

```python
# CHECKING DATA TYPES  

df.dtypes
```

```python
# cleaning data and adding column CAREER LENGTH


df['Rokie_Year'] = df['Span'].str.split("-").str[0]
df['Last_Year'] = df['Span'].str.split("-").str[1]
df = df.astype(
    {"Rokie_Year": 'int',
    'Last_Year':'int'
    }
)
df.drop(columns=['Span'],axis=1)
df['Career_Length'] = df['Last_Year'] - df['Rokie_Year']


```

```python
# CLEANING AND MANIPULATING BALL FACED COLUMNS 



#(DONE SPLITTING) df['Ball_Faced'] = df['Ball_Faced'].str.split('+').str[0]

#pd.set_option("display.max.rows",200)
df['Ball_Faced'] = df['Ball_Faced'].fillna(0)


```

```python
df = df.astype({
    'Ball_Faced':'int'
})
```

```python
# REMOVING * SIGN

#(DONE SPLITTING) df['Highest_Inns_Score'] = df['Highest_Inns_Score'].str.split('*').str[0]
df = df.astype({'Highest_Inns_Score':"int"})
```

```python

#AVERAGE CAREER LENGTH

df['Career_Length'].mean()
```

```python

df.dtypes
```

```python
#AVERAGE CAREER LENGTH THAT PLAY FOR MORE THAN 10 YEARS


more_than_10years = df[df['Career_Length'] > 10]
more_than_10years['Batting_Strike_Rate'].mean()
```

```python

# NUMBER OF PLAYERS THAT PLAYED BEFORE 1960
df[df['Rokie_Year'] < 1960]['Player'].count()
```

```python
#HIGHEST/MAX INNS SCORE PER COUNTRY

df.groupby('Country')['Highest_Inns_Score'].max().to_frame("HIGHINN_PER_COUNTRY").reset_index().sort_values('HIGHINN_PER_COUNTRY')
```
